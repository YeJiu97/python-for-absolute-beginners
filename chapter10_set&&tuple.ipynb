## set

set是具有唯一值的无序容器，而列表是有序的。

```python
my_set = set([1, 2, 3, 4, 5, 6, 6])

print(my_set)
```

可以发现set中的重复元素被自动去除了。

创建set的方法有两种：

1. 大括号 {}
2. set()

我们现在使用大括号来创建集合试试看：

```python
menu = {"apple", "banana", "orange"}

print(type(menu))
```

我们可以使用 - 来计算a集合中包括但是b集合中不包括的元素：

```python
a = set("12345")
b = set("123")

print(a-b)
```

我们可以使用 | 来获得集合a和集合b的合集：

```python
a = set("12345")
b = set("6789")

print(a|b)
```

## tuple

Python中的元组和列表类似，但是元组并不能够进行修改。

```python
tup1 = ("Google", "MateBook", "Alibaba")

type(tup1)
```

如果元组中只包含一个元素，则需要在最后添加,来进行表明，否则括号会被当成运算符来进行使用：

```python
tup2 = (50)

type(tup2)
```

```python
tup3 = (60,)
type(tup3)
```
