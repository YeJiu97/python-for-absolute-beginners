## 栈

栈（stack）是计算机科学中最基础的一种数据结构。

栈的核心原则是：先进后出（First In Last Out）。

如下图所示：



最先进入的数据会被压在最底部，直到最后才会被弹出来。

对于栈而言会有这样的几个常用名词：

1. 栈顶
2. 栈底
3. 进栈(push)
4. 出栈(pop)

## 程序实现

如此先入的4还留在栈中，而后入的7则被弹出来了。

我们可以创建一个叫做Stack的类：

```python
class Stack():
    
    def __init__(self):
        self.list_stack = []  # 初始化一个空列表
    
    # 用来和对list是否为空
    def is_empty(self):
        if not self.list_stack:  # not 0会得到True，否则都是False
            return True
        else:
            return False
    
    # 用来向栈追加一个项
    def push(self, item):
        self.list_stack.append(item)
    
    # 用来将栈的最后一项弹出 pop()默认弹出最后一个元素
    def pop(self):
        return self.list_stack.pop()
    
    # 检查stack是否为空
    def peek(self):
        if self.list_stack == []:  # 如果没有则没有返回值
            return None
        else:  
            return self.list_stack[-1]  # 我们可以查看最后一个被推入的元素
    
    def __repr__(self):
        return repr(self.list_stack)
```

```python
new_stack = Stack()
```

```python
new_stack.is_empty()
```

```python
new_stack.push(4)
```

```python
new_stack.peek()
```

```python
new_stack.push(0)
print(new_stack)
```

```python
new_stack.peek()
```

## 平衡括号

要求:会被要求输入一个由括号组成的字符串,括号会由各种括号组成,要求为判断这些括号的顺序是否是正确的,也就是是否是平衡的.

比如说:

1. [({})] 就是平衡的
2. {[}{] 就不是平衡的


```python
def balance_brackets(s):
    
    stack = []  # 创建一个空列表
    brackets = {'(':')', '[':']', '{':'}'}  # 用字典来创建对应关系
    
    for char in s:  # 开始对s中的每一个字符进行遍历
        
        # 这部分将会将s中的所有的左括号添
        if char in brackets.keys():  # keys()将会返回字典的所有的键
            stack.append(char)  # 将这个这个括号添加到stack的列表中
        
        # 如果不是左括号,那么就只有右括号了
        else:
            # 加入这个时候stack列表中还没有左括号
            # 又或者  stack中的最后一个左括号和右括号的映射的值不能够对上
            if len(stack) == 0 or brackets[stack.pop()] != char:
                return False  # 返回False
    
    return len(stack) == 0  # 如果堆栈的长度为0
```

```python
string_one = '[({})]'
balance_brackets(string_one)
```

```python
string_two = '{[}]'
balance_brackets(string_two)
```
