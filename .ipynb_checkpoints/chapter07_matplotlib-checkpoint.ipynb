## 随即漫步

简单来讲，基于过去的表现，无法预测将来的发展步骤和方向。

一个常见的例子就是：布朗运动。

````python
# 首先我们导入两个模块

import matplotlib.pyplot as plt
from random import choice
%matplotlib inline
````

````python
# 接着我们修改默认设定
plt.rc("figure", figsize = (12, 6))
````

````python
# 然后我们创建x和y的值
values = list(range(0, 55, 5))
values
````

````python
# 接着让我们将它们绘制出来
plt.plot(values)
plt.show()
````

````python
# 然后让我们创建随即漫步
step_option = [1, -1]
step_choice = choice(step_option)  # 从1和-1中随即选取
````

````python
walk = []  # 创建一个列表，这个列表会跟踪随即漫步的过程
walk.append(step_choice)  # 将第一步先加入进去

for step in range(1, 1000):
    next_step = walk[step-1] + choice(step_option)
    walk.append(next_step)  # 然后加入进去
````

````python
plt.plot(walk)
plt.show()
````

````python
# 我们可以将其打包成函数
def rand_walk(set_num):
    walk = []
    step_choice = choice([1, -1])
    walk.append(step_choice)
    
    for i in range(1, set_num):
        step_choice_2 = choice([1, -1])
        next_step = walk[i-1] + step_choice_2
        walk.append(next_step)
    return walk
````

````python
random_w = rand_walk(100)
print(len(random_w))
````

````python
plt.plot(walk)
plt.show()
````
