å¦‚æ­¤å…ˆå…¥çš„4è¿˜ç•™åœ¨æ ˆä¸­ï¼Œè€ŒåŽå…¥çš„7åˆ™è¢«å¼¹å‡ºæ¥äº†ã€‚

æˆ‘ä»¬å¯ä»¥åˆ›å»ºä¸€ä¸ªå«åšStackçš„ç±»ï¼š

```python
class Stack():
    
    def __init__(self):
        self.list_stack = []  # åˆå§‹åŒ–ä¸€ä¸ªç©ºåˆ—è¡¨
    
    # ç”¨æ¥å’Œå¯¹listæ˜¯å¦ä¸ºç©º
    def is_empty(self):
        if not self.list_stack:  # not 0ä¼šå¾—åˆ°Trueï¼Œå¦åˆ™éƒ½æ˜¯False
            return True
        else:
            return False
    
    # ç”¨æ¥å‘æ ˆè¿½åŠ ä¸€ä¸ªé¡¹
    def push(self, item):
        self.list_stack.append(item)
    
    # ç”¨æ¥å°†æ ˆçš„æœ€åŽä¸€é¡¹å¼¹å‡º pop()é»˜è®¤å¼¹å‡ºæœ€åŽä¸€ä¸ªå…ƒç´ 
    def pop(self):
        return self.list_stack.pop()
    
    # æ£€æŸ¥stackæ˜¯å¦ä¸ºç©º
    def peek(self):
        if self.list_stack == []:  # å¦‚æžœæ²¡æœ‰åˆ™æ²¡æœ‰è¿”å›žå€¼
            return None
        else:  
            return self.list_stack[-1]  # æˆ‘ä»¬å¯ä»¥æŸ¥çœ‹æœ€åŽä¸€ä¸ªè¢«æŽ¨å…¥çš„å…ƒç´ 
    
    def __repr__(self):
        return repr(self.list_stack)
```

```python
new_stack = Stack()
```

```python
new_stack.is_empty()
```

```python
new_stack.push(4)
```

```python
new_stack.peek()
```

```python
new_stack.push(0)
print(new_stack)
```

```python
new_stack.peek()
```

## å¹³è¡¡æ‹¬å·

è¦æ±‚:ä¼šè¢«è¦æ±‚è¾“å…¥ä¸€ä¸ªç”±æ‹¬å·ç»„æˆçš„å­—ç¬¦ä¸²,æ‹¬å·ä¼šç”±å„ç§æ‹¬å·ç»„æˆ,è¦æ±‚ä¸ºåˆ¤æ–­è¿™äº›æ‹¬å·çš„é¡ºåºæ˜¯å¦æ˜¯æ­£ç¡®çš„,ä¹Ÿå°±æ˜¯æ˜¯å¦æ˜¯å¹³è¡¡çš„.

æ¯”å¦‚è¯´:

1. [({})] å°±æ˜¯å¹³è¡¡çš„
2. {[}{] å°±ä¸æ˜¯å¹³è¡¡çš„


```python
def balance_brackets(s):
    
    stack = []  # åˆ›å»ºä¸€ä¸ªç©ºåˆ—è¡¨
    brackets = {'(':')', '[':']', '{':'}'}  # ç”¨å­—å…¸æ¥åˆ›å»ºå¯¹åº”å…³ç³»
    
    for char in s:  # å¼€å§‹å¯¹sä¸­çš„æ¯ä¸€ä¸ªå­—ç¬¦è¿›è¡ŒéåŽ†
        
        # è¿™éƒ¨åˆ†å°†ä¼šå°†sä¸­çš„æ‰€æœ‰çš„å·¦æ‹¬å·æ·»
        if char in brackets.keys():  # keys()å°†ä¼šè¿”å›žå­—å…¸çš„æ‰€æœ‰çš„é”®
            stack.append(char)  # å°†è¿™ä¸ªè¿™ä¸ªæ‹¬å·æ·»åŠ åˆ°stackçš„åˆ—è¡¨ä¸­
        
        # å¦‚æžœä¸æ˜¯å·¦æ‹¬å·,é‚£ä¹ˆå°±åªæœ‰å³æ‹¬å·äº†
        else:
            # åŠ å…¥è¿™ä¸ªæ—¶å€™stackåˆ—è¡¨ä¸­è¿˜æ²¡æœ‰å·¦æ‹¬å·
            # åˆæˆ–è€…  stackä¸­çš„æœ€åŽä¸€ä¸ªå·¦æ‹¬å·å’Œå³æ‹¬å·çš„æ˜ å°„çš„å€¼ä¸èƒ½å¤Ÿå¯¹ä¸Š
            if len(stack) == 0 or brackets[stack.pop()] != char:
                return False  # è¿”å›žFalse
    
    return len(stack) == 0  # å¦‚æžœå †æ ˆçš„é•¿åº¦ä¸º0
```

```python
string_one = '[({})]'
balance_brackets(string_one)
```

```python
string_two = '{[}]'
balance_brackets(string_two)
```
